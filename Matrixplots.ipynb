```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head(1)
```

```python
sns.heatmap(tips.corr(numeric_only=True))
```

```python
flights.head(2)
```

```python
pivot_data = flights.pivot_table(index="month", columns="year", values="passengers")
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(pivot_data, cmap='coolwarm', annot=True, fmt='.0f', linewidths=3, linecolor='black', annot_kws={'color':'black', 
                                                                                                            'fontsize':12,
                                                                                                            'fontweight':'bold'})

plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)
```

```python
plt.figure(figsize=(15,10))

sns.heatmap(
    pivot_data,
    cmap='coolwarm',
    annot=True,
    fmt='.0f',
    linewidths=2,
    linecolor='white',
    annot_kws={'size':12, 'weight':'bold'},
    cbar_kws={'shrink':0.8, 'label':'Value'}
)

plt.xticks(fontsize=14, rotation=45, ha='right')
plt.yticks(fontsize=14, rotation=360)
plt.title("Sales Heatmap", fontsize=18, fontweight='bold')
plt.tight_layout()
# plt.show()
```
