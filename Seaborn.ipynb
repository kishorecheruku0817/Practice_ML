```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel("/Users/kishorecheruku/python/Python/Data/Seaborn/supermarket_sales.xlsx")
mart.head()
```

Create the basic counter plot to get number of transactions for each product line and change the figure size

```python
# sns.countplot(data=mart, x=)
plt.figure(figsize=(10,8))
plt.xticks(rotation=45)
sns.countplot(data=mart, x='Product line')
```

Add hue to get the count on twocategories i.e Product line and gender

```python
plt.figure(figsize=(12,6))
plt.tight_layout()
plt.xticks(rotation=20)
sns.countplot(data=mart, x='Product line', hue='Gender')
```

```python
mart.head()
```

```python
plt.figure(figsize=(12,8))
# sns.barplot(mart.groupby('Product line')['Total'].sum(), width=0.5)
sns.barplot(data=mart, x='Product line', y='Total', hue='Gender' )
```

```python
plt.figure(figsize=(12,4))
ax = sns.barplot(data=mart, x='Product line', y='Total', hue='Gender', estimator=sum)

for container in ax.containers:
    ax.bar_label(container)
```

```python
plt.figure(figsize=(12,4))
agg_df = mart.groupby(['Product line', 'Gender'])['Total'].sum().reset_index()
bar = sns.barplot(data=agg_df, x='Product line', y='Total', hue='Gender')
print(agg_df)
for container in bar.containers:
    bar.bar_label(container)
```
