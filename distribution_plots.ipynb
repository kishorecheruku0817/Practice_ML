```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

Distplot

```python
sns.distplot(tips['total_bill'], kde=False, bins=20)
```

Jointplot

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='size')
```

pairplot

```python
sns.pairplot(tips, height=2, hue='sex', palette='coolwarm', kind='reg')
```

```python
sns.rugplot(data=tips, x='total_bill', y='tip')
```
