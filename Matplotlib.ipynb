```python
import matplotlib.pyplot as plt
data = [10, 34, 2, 5, 24, 8, 6]
weeks = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
avg = sum(data) / len(data)
plt.plot(weeks, data, label="Points", marker='o')
plt.axhline(avg, linestyle="--", color="orange", label="Average")
plt.xlabel("Weeks")
plt.ylabel("Data Points")
plt.legend()
plt.show()
```

```python
# Create temparature, weeks as x, y axes, show the average temparature and legend by using OO(Object oriented)
import matplotlib.pyplot as plt

temp = [13, 10, 6, -1, 5, 0, 4]
weeks = ["Mon", "Tue", "Wed", "Thur", "Fri", "Sat", "Sun"]
avg_temp = sum(temp) / len(temp)
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(weeks, temp, "r-", marker="o", label="Temp")
axes.set_xlabel("Weeks")
axes.set_ylabel("Temperatures")
axes.set_title("Boston")
axes.axhline(avg_temp, linestyle="--", color="green", label="Avg Temp")
axes.legend()
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.5, 0.3])
plt.show()
```

```python
humidity = [32, 33, 54, 21, 12, 46, 33]
fig, ax = plt.subplots(1, 2)

avg_temp = sum(temp) / len(temp)
avg_humidity = sum(humidity) / len(humidity)

# Left: Temperature
ax[0].plot(weeks, temp, marker="o", label="Temp")
ax[0].set_title("Boston - Temperature")
ax[0].set_xlabel("Weeks")
ax[0].set_ylabel("Temperature")
ax[0].axhline(avg_temp, label="Average Temp", linestyle="--")
ax[0].legend()

# Right: Humidity
ax[1].plot(weeks, humidity, marker=".", label="Humidity")
ax[1].set_title("Boston - Humidity")
ax[1].set_xlabel("Weeks")
ax[1].set_ylabel("Humidity")
ax[1].axhline(avg_humidity, label="Average Humidity", linestyle="--")
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import mplcursors

x = np.random.randint(0, 50, 20)
print(x)
y = np.random.randint(0, 50, 20)

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.2, 1, 1])
axes2 = fig.add_axes([0.5,0.3,0.3,0.3])

axes1.plot(x,y, 'go-', label='line 1', linewidth=2)
axes1.set_title("Larger Graph")

axes2.plot(y,x, marker='o', label='line 2')
axes2.set_title("Smaller Graph")
fig.legend()

mplcursors.cursor(axes1, hover=True)
mplcursors.cursor(axes2, hover=True)
plt.show()
```
