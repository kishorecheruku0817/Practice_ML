```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
def square_digit(d):
    return np.mean(d ** 2)

ax = sns.barplot(x='sex',y='total_bill',data=tips, estimator=square_digit, color="b")
for container in ax.containers:
    ax.bar_label(container)
```

```python
ax = sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
for i, artist in enumerate(ax.artists):
    # Add median values or other statistics as text
    # This is just an example - you'd need to calculate the actual values
    ax.annotate(f"Median", xy=(i, 0), xytext=(0, 10), 
               textcoords="offset points", ha='center')
```

```python
ax = sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
for container in ax.containers:
    ax.bar_label(container)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue="sex")
sns.violinplot(x='day', y='total_bill', data=tips)
```
