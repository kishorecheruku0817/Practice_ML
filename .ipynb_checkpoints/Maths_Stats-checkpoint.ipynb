```python
import numpy as np
from scipy import stats
data = np.array([5, 7, 7, 9, 10])
print(np.mean(data))
print(np.median(data))
print(stats.mode(data))
```

```python
import numpy as np
import statistics as stats
data = np.array([12, 15, 15, 15, 18, 18, 18, 20, 22])
mean = np.mean(data)
median = np.median(data)
print(stats.multimode(data))

if mean > median:
    print("Right Skewed")
elif mean < median:
    print("Left Skewed")
else:
    print("Symmetric")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = np.array([40, 42, 39, 41, 40, 300, 38, 41, 40])

# Compute IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

outliers = data[(data < lower) | (data > upper)]

# Plot
plt.figure(figsize=(6, 3))
sns.boxplot(x=data)

# Annotate outliers
# for val in outliers:
#     plt.text(
#         x=0,
#         y=val,
#         s=str(val),
#         color="r",
#         ha="center",
#         va="top",
#         fontsize=10,
#         fontweight="bold"
#     )

plt.title("Boxplot with Outlier Values")
plt.show()
```
