```python
def correlation(list1, list2):
    mean1 = list1.mean()
    mean2 = list2.mean()
    numerator = 0
    sum_x_sq = 0
    sum_y_sq = 0
    for i in range(0, len(list1)):
        x = list1[i] - mean1
        y = list2[i] - mean2
        
        numerator += x * y            #  (x - mean_x)(y - mean_y)
        sum_x_sq += x ** 2            #  (x - mean_x)^2
        sum_y_sq += y ** 2            #  (y - mean_y)^2
    
    denominator = (sum_x_sq * sum_y_sq) ** 0.5
    r = numerator / denominator
    print(r)
        
```

```python
import pandas as pd

hours = pd.Series([40,45,35,50,30,42])
scores = pd.Series([88,92,80,95,78,85])

correlation(hours, scores)
```
